```python
import os
import numpy as np
import pandas as pd
import pickle as pkl

import matplotlib.pyplot as plt
```

### 'Abortion' data set

```python
# load the cluster distance
with open("../pickle/abortion_daily_between.pkl", "rb") as file:
    k_between, pickle_idx = pkl.load(file)
    
# load the indexes corresponding to the non-zero entries
with open("../pickle/abortion_daily_nonzero_indexes.pkl", "rb") as f:
    nonzero_idx = pkl.load(f)
    
nonzero_len = []
for i in nonzero_idx[980:]:
    nonzero_len.append(len(i))
    
k_between = np.array(k_between[980:])
```

```python
# load intra-cluster distances
with open("../pickle/abortion_daily_intra_full.pkl", "rb") as file:
    k0, k1, ki = pkl.load(file)
```

```python
idx = np.linspace(0, len(nonzero_len), 11)
idx = idx[1:]
idx = np.round(idx)
```

```python
mean_ = []
std_  = []

_arr = []

it = 0
for i, j in sorted(zip(nonzero_len, k_between)):
    _arr.append(j)
    if it in idx:
        mean_.append(np.array(_arr).mean())
        std_.append(np.array(_arr).std())
        _arr = []
    it += 1

mean_.append(np.array(_arr).mean())
std_.append(np.array(_arr).std())
    
```

```python
plt.figure(figsize = (6, 3.5))
plt.rcParams.update({'font.size': 16})
plt.errorbar(range(0, 10), mean_, std_, linewidth = 2, capsize = 10, linestyle = '--')
plt.xlabel("Bucket No.")
plt.ylabel("Inter-cluster distance")
plt.title("Mean value ('abortion')")
plt.xticks(ticks = range(0, 10), labels = range(1, 11))
plt.grid()

plt.savefig("../plots/ad_hoc/_ab_inter_cluster_buckets.png", bbox_inches = 'tight')
```

```python
# intra values
mean_ = []
std_  = []

_arr = []

it = 0
for i, j in sorted(zip(nonzero_len, k0)):
    _arr.append(j)
    if it in idx:
        mean_.append(np.array(_arr).mean())
        std_.append(np.array(_arr).std())
        _arr = []
    it += 1

mean_.append(np.array(_arr).mean())
std_.append(np.array(_arr).std())
```

```python
plt.figure(figsize = (6, 3.5))
plt.rcParams.update({'font.size': 16})
plt.errorbar(range(0, 10), mean_, std_, linewidth = 2, capsize = 10, linestyle = '--')
plt.xlabel("Bucket No.")
plt.ylabel("Intra-cluster distance")
plt.title("Mean value ('abortion')")
plt.xticks(ticks = range(0, 10), labels = range(1, 11))
plt.grid()

plt.savefig("../plots/ad_hoc/_ab_intra_d_cluster_buckets.png", bbox_inches = 'tight')
```
