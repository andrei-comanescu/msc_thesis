```python
import os
import numpy as np
import pandas as pd
import pickle as pkl

import matplotlib.pyplot as plt
```

## 'Abortion' data set

```python
# load the indexes corresponding to the non-zero entries
with open("../pickle/abortion_daily_nonzero_indexes.pkl", "rb") as f:
    nonzero_idx = pkl.load(f)
    
# load the correlations and the percentages
with open("../pickle/abortion_daily_active_users_corr.pkl", "rb") as f:
    corr_agg, perc_d_agg, perc_r_agg = pkl.load(f)

# load the correlation rings and the percentages
with open("../pickle/abortion_daily_active_users_corr_rings.pkl", "rb") as f:
    corr_rings, perc_d_rings, perc_r_rings = pkl.load(f)
```

```python
nonzero_len = []
for i in nonzero_idx[980:]:
    nonzero_len.append(len(i))
    
corr_agg   = np.array(corr_agg)
perc_d_agg = np.array(perc_d_agg)
perc_r_agg = np.array(perc_r_agg)

corr_rings   = np.array(corr_rings)
perc_d_rings = np.array(perc_d_rings)
perc_r_rings = np.array(perc_r_rings)
```

```python
idx = np.linspace(0, 1805, 11)
idx = idx[1:]
idx = np.round(idx)
```

```python
# correlation
mean_agg = []
std_agg  = []
for k in range(0, len(corr_agg)):
    mean_corr = []
    std_corr  = []

    _arr = []

    it  = 0
    for i, j in sorted(zip(nonzero_len, corr_agg[k])):
        _arr.append(j)
        if it in idx:
            mean_corr.append(np.array(_arr).mean())
            std_corr.append(np.array(_arr).std())
            _arr = []
        it += 1

    mean_corr.append(np.array(_arr).mean())
    std_corr.append(np.array(_arr).std())
    
    mean_agg.append(mean_corr)
    std_agg.append(std_corr)
    
# percentage democrats
mean_d = []
std_d  = []
for k in range(0, len(perc_d_agg)):
    mean_corr = []
    std_corr  = []

    _arr = []

    it  = 0
    for i, j in sorted(zip(nonzero_len, perc_d_agg[k])):
        _arr.append(j)
        if it in idx:
            mean_corr.append(np.array(_arr).mean())
            std_corr.append(np.array(_arr).std())
            _arr = []
        it += 1

    mean_corr.append(np.array(_arr).mean())
    std_corr.append(np.array(_arr).std())
    
    mean_d.append(mean_corr)
    std_d.append(std_corr)
    
# percentage republicans
mean_r = []
std_r  = []
for k in range(0, len(perc_r_agg)):
    mean_corr = []
    std_corr  = []

    _arr = []

    it  = 0
    for i, j in sorted(zip(nonzero_len, perc_r_agg[k])):
        _arr.append(j)
        if it in idx:
            mean_corr.append(np.array(_arr).mean())
            std_corr.append(np.array(_arr).std())
            _arr = []
        it += 1

    mean_corr.append(np.array(_arr).mean())
    std_corr.append(np.array(_arr).std())
    
    mean_r.append(mean_corr)
    std_r.append(std_corr)
    
# correlation rings
mean_agg_rings = []
std_agg_rings  = []
for k in range(0, len(corr_rings)):
    mean_corr = []
    std_corr  = []

    _arr = []

    it  = 0
    for i, j in sorted(zip(nonzero_len, corr_rings[k])):
        _arr.append(j)
        if it in idx:
            mean_corr.append(np.array(_arr).mean())
            std_corr.append(np.array(_arr).std())
            _arr = []
        it += 1

    mean_corr.append(np.array(_arr).mean())
    std_corr.append(np.array(_arr).std())
    
    mean_agg_rings.append(mean_corr)
    std_agg_rings.append(std_corr)
    
# percentage rings democrats
mean_d_rings = []
std_d_rings  = []
for k in range(0, len(perc_d_rings)):
    mean_corr = []
    std_corr  = []

    _arr = []

    it  = 0
    for i, j in sorted(zip(nonzero_len, perc_d_rings[k])):
        _arr.append(j)
        if it in idx:
            mean_corr.append(np.array(_arr).mean())
            std_corr.append(np.array(_arr).std())
            _arr = []
        it += 1

    mean_corr.append(np.array(_arr).mean())
    std_corr.append(np.array(_arr).std())
    
    mean_d_rings.append(mean_corr)
    std_d_rings.append(std_corr)
    
# percentage rings republicans
mean_r_rings = []
std_r_rings  = []
for k in range(0, len(perc_r_rings)):
    mean_corr = []
    std_corr  = []

    _arr = []

    it  = 0
    for i, j in sorted(zip(nonzero_len, perc_r_rings[k])):
        _arr.append(j)
        if it in idx:
            mean_corr.append(np.array(_arr).mean())
            std_corr.append(np.array(_arr).std())
            _arr = []
        it += 1

    mean_corr.append(np.array(_arr).mean())
    std_corr.append(np.array(_arr).std())
    
    mean_r_rings.append(mean_corr)
    std_r_rings.append(std_corr)
    
# get them as np.arrays
mean_agg = np.array(mean_agg)
std_agg  = np.array(std_agg)

mean_d = np.array(mean_d)
std_d  = np.array(std_d)

mean_r = np.array(mean_r)
std_r  = np.array(std_r)

mean_agg_rings = np.array(mean_agg_rings)
std_agg_rings  = np.array(std_agg_rings)

mean_d_rings = np.array(mean_d_rings)
std_d_rings  = np.array(std_d_rings)

mean_r_rings = np.array(mean_r_rings)
std_r_rings  = np.array(std_r_rings)
```

```python
# correlation
mean_agg = []
std_agg  = []
for k in range(0, len(corr_agg)):
    mean_corr = []
    std_corr  = []

    _arr = []

    it  = 0
    for i, j in sorted(zip(nonzero_len, corr_agg[k])):
        if ~np.isnan(j):
            _arr.append(j)
        if it in idx:
            mean_corr.append(np.array(_arr).mean())
            std_corr.append(np.array(_arr).std())
            _arr = []
        it += 1

    mean_corr.append(np.array(_arr).mean())
    std_corr.append(np.array(_arr).std())
    
    mean_agg.append(mean_corr)
    std_agg.append(std_corr)
    
# percentage democrats
mean_d = []
std_d  = []
for k in range(0, len(perc_d_agg)):
    mean_corr = []
    std_corr  = []

    _arr = []

    it  = 0
    for i, j in sorted(zip(nonzero_len, perc_d_agg[k])):
        if ~np.isnan(j):
            _arr.append(j)
        if it in idx:
            mean_corr.append(np.array(_arr).mean())
            std_corr.append(np.array(_arr).std())
            _arr = []
        it += 1

    mean_corr.append(np.array(_arr).mean())
    std_corr.append(np.array(_arr).std())
    
    mean_d.append(mean_corr)
    std_d.append(std_corr)
    
# percentage republicans
mean_r = []
std_r  = []
for k in range(0, len(perc_r_agg)):
    mean_corr = []
    std_corr  = []

    _arr = []

    it  = 0
    for i, j in sorted(zip(nonzero_len, perc_r_agg[k])):
        if ~np.isnan(j):
            _arr.append(j)
        if it in idx:
            mean_corr.append(np.array(_arr).mean())
            std_corr.append(np.array(_arr).std())
            _arr = []
        it += 1

    mean_corr.append(np.array(_arr).mean())
    std_corr.append(np.array(_arr).std())
    
    mean_r.append(mean_corr)
    std_r.append(std_corr)
    
# correlation rings
mean_agg_rings = []
std_agg_rings  = []
for k in range(0, len(corr_rings)):
    mean_corr = []
    std_corr  = []

    _arr = []

    it  = 0
    for i, j in sorted(zip(nonzero_len, corr_rings[k])):
        if ~np.isnan(j):
            _arr.append(j)
        if it in idx:
            mean_corr.append(np.array(_arr).mean())
            std_corr.append(np.array(_arr).std())
            _arr = []
        it += 1

    mean_corr.append(np.array(_arr).mean())
    std_corr.append(np.array(_arr).std())
    
    mean_agg_rings.append(mean_corr)
    std_agg_rings.append(std_corr)
    
# percentage rings democrats
mean_d_rings = []
std_d_rings  = []
for k in range(0, len(perc_d_rings)):
    mean_corr = []
    std_corr  = []

    _arr = []

    it  = 0
    for i, j in sorted(zip(nonzero_len, perc_d_rings[k])):
        if ~np.isnan(j):
            _arr.append(j)
        if it in idx:
            mean_corr.append(np.array(_arr).mean())
            std_corr.append(np.array(_arr).std())
            _arr = []
        it += 1

    mean_corr.append(np.array(_arr).mean())
    std_corr.append(np.array(_arr).std())
    
    mean_d_rings.append(mean_corr)
    std_d_rings.append(std_corr)
    
# percentage rings republicans
mean_r_rings = []
std_r_rings  = []
for k in range(0, len(perc_r_rings)):
    mean_corr = []
    std_corr  = []

    _arr = []

    it  = 0
    for i, j in sorted(zip(nonzero_len, perc_r_rings[k])):
        if ~np.isnan(j):
            _arr.append(j)
        if it in idx:
            mean_corr.append(np.array(_arr).mean())
            std_corr.append(np.array(_arr).std())
            _arr = []
        it += 1

    mean_corr.append(np.array(_arr).mean())
    std_corr.append(np.array(_arr).std())
    
    mean_r_rings.append(mean_corr)
    std_r_rings.append(std_corr)
    
# get them as np.arrays
mean_agg = np.array(mean_agg)
std_agg  = np.array(std_agg)

mean_d = np.array(mean_d)
std_d  = np.array(std_d)

mean_r = np.array(mean_r)
std_r  = np.array(std_r)

mean_agg_rings = np.array(mean_agg_rings)
std_agg_rings  = np.array(std_agg_rings)

mean_d_rings = np.array(mean_d_rings)
std_d_rings  = np.array(std_d_rings)

mean_r_rings = np.array(mean_r_rings)
std_r_rings  = np.array(std_r_rings)
```
