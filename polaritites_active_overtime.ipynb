```python
import os
import numpy as np
import pandas as pd
import pickle as pkl

import matplotlib.pyplot as plt
```

### 'Abortion' data set

```python
# load the indexes corresponding to the non-zero entries
with open("../pickle/abortion_daily_nonzero_indexes.pkl", "rb") as f:
    nonzero_idx = pkl.load(f)
    
nonzero_idx = np.array(nonzero_idx[980:])
    
# get the polarity scores
polarities = pd.read_csv("../keras/abortion_daily/ordered_predictions.csv")
```

```python
rep_perc = []
for i in nonzero_idx:
    rep_perc.append(100 * polarities.loc[i, 'Class'].sum() / len(i))
```

```python
nonzero_len = []
for i in nonzero_idx:
    nonzero_len.append(len(i))
    
idx = np.linspace(0, len(nonzero_len), 11)
idx = idx[1:]
idx = np.round(idx)
```

```python
mean_ = []
std_  = []

_arr = []

it = 0
for i, j in sorted(zip(nonzero_len, rep_perc)):
    _arr.append(j)
    if it in idx:
        mean_.append(np.array(_arr).mean())
        std_.append(np.array(_arr).std())
        _arr = []
    it += 1

mean_.append(np.array(_arr).mean())
std_.append(np.array(_arr).std())
    
```

```python
plt.figure(figsize = (6, 3.5))
plt.rcParams.update({'font.size': 18})
plt.errorbar(range(0, 10), mean_, std_, linewidth = 2.2, capsize = 10, linestyle = '--')
plt.xlabel("Bucket Number")
plt.ylabel("Percentage")
plt.title("Republicans ('abortion')")
plt.xticks(ticks = range(0, 10), labels = range(1, 11))
plt.grid()

plt.savefig("../plots/ad_hoc/_ab_rep_active_buckets.png", bbox_inches = 'tight')
```
