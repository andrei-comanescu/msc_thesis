```python
import os
import numpy as np
import pickle as pkl

import matplotlib.pyplot as plt
```

## Active users for the 'abortion' data set

```python
vals = np.load("../npy/abortion_daily/node_degree.npy")

v       = []
sums    = []
indexes = []
for i in range(0, vals.shape[1]):
    v.append(len(np.flatnonzero(vals[:, i])))
    sums.append(vals[np.flatnonzero(vals[:, i]), i].sum())
    indexes.append(np.flatnonzero(vals[:, i]))
```

```python
# save the indexes corresponding to the non-zero entries
with open("../pickle/abortion_daily_nonzero_indexes.pkl", "wb") as f:
    pkl.dump(indexes, f)
```

```python
# get dates mapping
pickle_folder_path = "../pickle/abortion_daily_largest_aligned_norm2/"

pickle_files_aligned = [f for f in os.listdir(pickle_folder_path) if os.path.isfile(os.path.join(pickle_folder_path, f))]

pickle_idx = []
for pickle_ in pickle_files_aligned:
    pickle_idx.append((pickle_.split(".")[0].split("_")[1]))

pickle_idx.sort()
```

```python
idx  = []
pidx = []
for i in range(980, len(v), 250): 
    idx.append(i)
    pidx.append(i)
```

```python
plt.figure(figsize = (20, 4))
plt.rcParams.update({'font.size': 16})
plt.bar(x = range(980, len(v)), height = v[980:], color = 'k')
plt.plot(pickle_idx.index('2012-10-11') * np.ones(2), [0, 9000], linestyle = '--', linewidth = 1.5, color = 'r', alpha = 0.8)
plt.plot(pickle_idx.index('2013-06-11') * np.ones(2), [0, 9000], linestyle = '--', linewidth = 1.5, color = 'r', alpha = 0.8)
plt.plot(pickle_idx.index('2013-07-18') * np.ones(2), [0, 9000], linestyle = '--', linewidth = 1.5, color = 'r', alpha = 0.8)
plt.plot(pickle_idx.index('2015-07-14') * np.ones(2), [0, 9000], linestyle = '--', linewidth = 1.5, color = 'r', alpha = 0.8)
plt.plot(pickle_idx.index('2015-11-27') * np.ones(2), [0, 9000], linestyle = '--', linewidth = 1.5, color = 'r', alpha = 0.8)
plt.plot(pickle_idx.index('2016-06-27') * np.ones(2), [0, 9000], linestyle = '--', linewidth = 1.5, color = 'r', alpha = 0.8)
plt.ylim([0, 9000])
plt.xticks(ticks = idx, labels = np.array(pickle_idx)[pidx])
plt.xlabel("Date")
plt.ylabel("Count")
plt.title("Unique users per day ('abortion' data set)")
plt.grid()

plt.savefig("../plots/ad_hoc/_ab_users.png", bbox_inches = 'tight')
```

```python
plt.figure(figsize = (20, 4))
plt.rcParams.update({'font.size': 14})
plt.bar(x = range(980, len(sums)), height = sums[980:], color = 'k', alpha = 1)
plt.plot(pickle_idx.index('2012-10-11') * np.ones(2), [0, 40000], linewidth = 1.5, color = 'b', alpha = 0.5)
plt.plot(pickle_idx.index('2013-06-11') * np.ones(2), [0, 40000], linewidth = 1.5, color = 'b', alpha = 0.5)
plt.plot(pickle_idx.index('2013-07-18') * np.ones(2), [0, 40000], linewidth = 1.5, color = 'b', alpha = 0.5)
plt.plot(pickle_idx.index('2015-07-14') * np.ones(2), [0, 40000], linewidth = 1.5, color = 'b', alpha = 0.5)
plt.plot(pickle_idx.index('2015-11-27') * np.ones(2), [0, 40000], linewidth = 1.5, color = 'b', alpha = 0.5)
plt.plot(pickle_idx.index('2016-06-27') * np.ones(2), [0, 40000], linewidth = 1.5, color = 'b', alpha = 0.5)
plt.ylim([0, 40000])
plt.xticks(ticks = idx, labels = np.array(pickle_idx)[pidx])
plt.xlabel("Date")
plt.ylabel("Count")
plt.title("Conversations per day ('abortion' data set)")
plt.grid()

plt.savefig("../plots/abortion_daily/conversations.png", bbox_inches = 'tight')
```
